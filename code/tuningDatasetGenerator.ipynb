```python
import pandas as pd
import random

# Расширенный список ответов на вопрос "Что бы вы хотели видеть на следующем фуршете?"
answers = [
    "вода", "водка", "мясные блюда", "мясо", "рыба", "суши", "салаты", "соки", "алкоголь", "вино", "шампанское",
    "фрукты", "овощи", "сладости", "десерты", "торты", "чипсы", "кексы", "пирожные", "хлеб", "блины", "шашлыки", 
    "пицца", "гамбургеры", "бутерброды", "сухофрукты", "орехи", "сыр", "колбаса", "пельмени", "мороженое", "смузи", 
    "кофе", "чай", "коктейли", "минеральная вода", "газированная вода", "пироги", "шоколад", "зефир", "мармелад", 
    "картофель фри", "гриль", "сэндвичи", "пастила", "яблоки", "груши", "виноград", "персики", "абрикосы", "сок гранатовый",
    "сок апельсиновый", "мандариновый сок", "ананасовый сок", "выпечка", "торт", "бисквит", "варенье", "джем", "восточные сладости",
    "хумус", "лаваш", "мясо на гриле", "крылышки", "ребрышки", "супы", "борщ", "уха", "окрошка", "холодец", "плов", "курица",
    "индейка", "тирамису", "медовик", "чизкейк", "эклеры", "тарталетки", "роллы", "спагетти", "паста", "картофель", "овощной салат",
    "фруктовый салат", "винегрет", "оливье", "грибы", "баклажаны", "кабачки", "арбузы", "дыни", "йогурты", "творог", "сметана", 
    "красная икра", "черная икра", "рулеты", "фаршированная рыба", "мед", "конфеты", "пряники", "картофельное пюре"
]

# Примеры схожих пар
similar_pairs = [
    ("мясные блюда", "мясо"), ("сладости", "десерты"), ("вино", "алкоголь"), ("напитки", "соки"), 
    ("фрукты", "овощи"), ("мясо на гриле", "шашлыки"), ("пицца", "гамбургеры"), ("кофе", "чай"),
    ("чизкейк", "медовик"), ("смузи", "коктейли"), ("пельмени", "шашлыки"), ("суши", "роллы"),
    ("овощной салат", "фруктовый салат"), ("борщ", "супы"), ("яблоки", "груши"), ("торт", "пирог"),
    ("газированная вода", "минеральная вода"), ("кабачки", "баклажаны"), ("арбуз", "дыня"),
    ("сметана", "творог"), ("творог", "йогурты"), ("оливье", "винегрет")
]

# Генерация случайных пар
def generate_random_pairs(answers, similar_pairs, num_pairs=300):
    data = []
    for _ in range(num_pairs):
        a1, a2 = random.sample(answers, 2)
        similarity = 1 if (a1, a2) in similar_pairs or (a2, a1) in similar_pairs else 0
        data.append([a1, a2, similarity])
    return data

# Создаем датасет
random_pairs = generate_random_pairs(answers, similar_pairs, num_pairs=300)

# Сохраняем как DataFrame и в CSV файл
df = pd.DataFrame(random_pairs, columns=["Ответ 1", "Ответ 2", "Схожесть"])
df.to_csv("../data/ftDataset.csv", index=False)

print(df.head(10))
```
